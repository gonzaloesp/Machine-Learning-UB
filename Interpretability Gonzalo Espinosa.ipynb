## Interpretability is twofold

David has introduced the concept of machine learning interpretability as the ability of a given human to understand why the model is behaving like it is doing. Understanding what is the set of attributes that the model used and what are the coefficients associated with each of those attribute in the model might be a sufficient condition for interpreting the results. Visualizing the results, for example through decision boundaries, is key for model interpretation.

From my point of view, interpretability can be divided into two varieties [1]: **transparency** and **post-hoc**. 

### Transparency

Transparency is the concept that David has explained that is the ability of introspecting the model and determining precisely what it's doing. A transparent model should help you understand how the factors that are included in the model led to a prediction and should allow you to adjust how you use the prediction based on these additional factors. 

Some publications suggest that the most useful models aren't especially transparent, but in my opinion all models are as transparent as the will we have to understand it. Some models are more complex than others, but all models are understandable. If one understands the model class, the loss and the learning algorithm, then has the control over the problem. The point is that the majority machine learning practitioners donâ€™t care about how models work and they usually only care about predictive performance through evaluation metrics, but this is another discussion. 

However, is true that not always is easy to visualize decision boundaries because we are only able to observe till 3-dimensional plots. In those cases, in which we have a lot of features, we only can visualize how classifiers predict respect some limited number of features (2-3) at the same time. Nevertheless, in case that we could visualize decision boundaries, visualize is equally feasible for all models, from the simplest model KNN to the most complex AdaBoost. ### Post-hoc

On the other hand, the idea of post-hoc address to the ability to extract conclusions and generalize it if applicable. This appeal to the fact that our models can undercover insights from our data and can explain real phenomena. From a statistical point of view, one example of post-hoc analysis is the inference on parameters. For instance, in linear regression we can interpret the value of its coefficients like the dependency of dependent variables with the target. Furthermore, models allow drawing conclusions even the presence of uncertainty, for example, through confidence intervals or credible intervals (depending on the statistical approach) of parameters of models. 

However, not all models are capable to make inferences. For example, in a decision tree we can understand how instances are classified through decision rules, but we canâ€™t generalize that these decision rules are â€œuniversalâ€ and always true because two equally valid (good classifier) trees could arrive different or even contrary rules about data. In the sense of post-hoc analysis is true that some models are more interpretable than others.

To sum up, the idea behind interpretability is that in addition to the predictions required by our evaluation metrics, models should yield some additional information: interpretation. Through interpretation, we can understand not only how model works, but also obtain insights of our data.
